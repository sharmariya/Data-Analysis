# Customer Age-wise Distribution

```python
# Create a new figure and axis with a specified size (20 units wide and 5 units tall)
fig, age_dis = plt.subplots(figsize=(20, 5))

# Create a histogram of the 'Age' column from the DataFrame 'df'
# with 20 bins(20 bars of histogram), black edgecolor, 50% transparency (alpha), and purple color.
# The 'density' parameter is set to True to represent the data as a probability density.
age_dis.hist(df['Age'], bins=20, edgecolor='black', alpha=0.5, color='purple', density=True)

# Create a kernel density estimate (KDE) plot for the 'Age' column from the DataFrame 'df'in yellow color and overlay it on the same axis 'age_dis'.
df['Age'].plot(kind='kde', color='yellow', ax=age_dis)

# Set the label for the x-axis to 'Age'
age_dis.set_xlabel('Age')

# Set the label for the y-axis to 'Count'
age_dis.set_ylabel('Count')

# Add a legend to the plot with labels 'Density Curve' and 'Histogram'
age_dis.legend(['Density Curve', 'Histogram'])

# Set the title of the plot to 'Customer Age-wise Distribution'
age_dis.set_title('Customer Age-wise Distribution')

# Display the plot
plt.show()
```

# Gender-wise distribution of customers

```python
# Create a new figure for the pie chart with a specific size (20 units wide and 5 units tall)
plt.figure(figsize=(20, 5))
 
# Count the occurrences of each unique value in the "Gender" column of the DataFrame 'df'
count = df["Gender"].value_counts()

# Define the 'explode' parameter to create a slight separation between pie chart segments
explode = (0, 0.05)

# Create a pie chart using the counted values from the "Gender" column
# Set the font size to 20, specify custom colors, use the 'explode' parameter for exploading the pie chart,
# and using autopct='%1.1f%%' to display percentages with one decimal place
count.plot(kind='pie', fontsize=15, colors=["skyblue","Pink"], explode=explode, autopct='%1.1f%%')

# Set the label for the x-axis to 'Gender' which is written in bold and in green color and font size is 15
plt.xlabel('Gender', weight="bold", color="Green", fontsize=15)

# Ensure the pie chart is circular by setting the aspect ratio to 'equal'
plt.axis('equal')

# Add a legend to the pie chart using the unique values from the "Gender" column
# and place it at the best location on the chart
plt.legend(labels=count.index, loc="best")

# Display the pie chart
plt.show()
```

# Frequency of Purchases

```python
# Create a new figure with a specific size (20 units in width and 5 units in height)
plt.figure(figsize=(20, 5))

# Extract the "Frequency of Purchases" column from the DataFrame, count the values, and create a bar plot.
# The bar colors are specified, and the x-axis labels are not rotated (rot = 0).
fp = df["Frequency of Purchases"].value_counts().plot(kind='bar', color=["yellow", "green", "red", "skyblue", "orange", "pink", "brown"], rot=0)

# Loop through each bar patch in the bar plot.
for p in fp.patches:
    # The code iterates through each bar in a bar plot and adds a text label slightly above each bar's center and slightly above the top of the bar.
    #The label is the height (count) of the bar, and it's centered horizontally and anchored to the bottom of the text. 
    #The adjustments of 0.25 to the x-coordinate (horizontal) and 1 to the y-coordinate (vertical) ensure 
    #that the annotations are appropriately positioned for readability and appearance. 
    #This annotates the plot with the counts for each bar.
    fp.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha='center', va='bottom', color='black')

# Set the tick labels' font size for both x and y axes to 15.
fp.tick_params(axis='both', labelsize=15)

# Set the label for the x-axis with custom styling (bold, black color, and fontsize 15).
plt.xlabel('Frequency of Purchases', weight="bold", color="black", fontsize=15)

# Set the label for the y-axis with custom styling (bold, black color, and fontsize 15).
plt.ylabel('Count', weight="bold", color="black", fontsize=15);
```

# Category-wise Distribution of Items Sold

```python
# Create a new figure for the pie chart with a specific size of 20x5 inches
plt.figure(figsize=(20, 5))

# Calculate the counts of each category in the DataFrame's "Category" column
ct = df["Category"].value_counts()

# Define the "explode" values to separate slices in the pie chart (for visual effect)
explode = (0.05, 0.1, 0.15, 0.2)
 
# Create a pie chart using the counted values from the "Category" column
# Set the font size to 15, specify custom colors, use the 'explode' parameter for exploading the pie chart,
# and using autopct='%1.2f%%' to display percentages upto 2 decimal places

ct.plot(kind='pie', fontsize=15, colors=["yellow", "green", "red", "skyblue", "pink"], explode=explode, autopct='%1.2f%%')

# Add a label for the x-axis (Categories), bold and brown in color with font size 15
plt.xlabel('Categories', weight="bold", color="brown", fontsize=15)

# Set the aspect ratio of the pie chart to be equal, ensuring it appears as a circle
plt.axis('equal')

# Add a legend to the pie chart using category labels and placing it in the "best" location
plt.legend(labels=ct.index, loc="best")

# Display the pie chart
plt.show()
```

# Payment Methods

```python
# Create a new figure with a specified figure size (25 units wide and 7 units tall)
plt.figure(figsize=(25, 7))

# Calculate the value counts of the "Payment Method" column in the DataFrame 'df' and create a bar plot
# The bars are colored with a specified list of colors, and the x-axis labels are not rotated (rot=0)
pm = df["Payment Method"].value_counts().plot(kind='bar', color=["yellow", "green", "red", "skyblue", "orange", "pink"], rot=0)

# Iterate through the bar patches to annotate each bar with its count value
for i in pm.patches:
    # Annotate the bar with an integer representing its height
    # The annotation is centered above the bar, slightly shifted horizontally and vertically
    # The text color is set to black
    pm.annotate(int(i.get_height()), (i.get_x() + 0.25, i.get_height() + 1), ha='center', va='bottom', color='black')
    
# Adjust the tick label size for both x and y axes to 15
pm.tick_params(axis='both', labelsize=15)

# Set the label for the x-axis with custom properties: bold, blue text, and a fontsize of 20
plt.xlabel('Payment Method', weight="bold", color="blue", fontsize=20)

# Set the label for the y-axis with custom properties: bold, blue text, and a fontsize of 20
plt.ylabel('Count', weight="bold", color="blue", fontsize=20)
```

# Maximum and Minimum Review Rating

```python
max_r = df['Review Rating'].max() #finds the maximum review rating 
min_r = df['Review Rating'].min() #finds the min review rating
print("Maximum Review Rating is:", max_r)
print("Minimum Review Rating is:", min_r)
```

# The Most Common Item Purchased

```python
ab = df['Item Purchased'].mode()[0]
#The .mode() function is used to calculate the mode of the column, and [0] is used to access the first mode value.

print("Most Common Item Purchased is:", ab)
```

# Average Review Rating by Male and Female Customers

```python
m = df[df['Gender'] == 'Male']['Review Rating'].mean() #calculates the mean review rating when gender is male
f = df[df['Gender'] == 'Female']['Review Rating'].mean() #calculates the mean review rating when gender is female
print("Average Review Rating by Male Customers:", m)
print("Average Review Rating by Female Customers:",f)
```

# The Most Common Season for Purchases

```python
season = df['Season'].mode()[0]

# Using the 'mode()' function to find the most common (mode) value in the 'Season' column.
# The result of 'mode()' in a Pandas Series with one or more mode values.
# So to access the first mode value, we access it with '[0]'.

print("Most Common Season for Purchases is", season)
```

# Total Purchase Amount for Each Size of Clothing Items

```python
ps = df[df['Category'] == 'Clothing'].groupby('Size')['Purchase Amount (USD)'].sum()

#It filters the DataFrame 'df' to select rows where the 'Category' column is equal to 'Clothing'.
# It groups the filtered DataFrame by the 'Size' column.
# It then calculates the sum of the 'Purchase Amount (USD)' column for each group.

print("Total Purchase Amount by Size for Clothing Items are as follows:")
print(ps)
```

# The Total Purchase Amount for Customers who made Purchases in the Winter Season and used PayPal as the Payment Method

```python
amt = df[(df['Season'] == 'Winter') & (df['Payment Method'] == 'PayPal')]['Purchase Amount (USD)'].sum()
#filters the DataFrame to select rows where both 'Season' is 'Winter' and 'Payment Method' is 'PayPal',
#and then calculates the sum of the 'Purchase Amount (USD)' for these filtered rows. 
#The result is stored in the variable 'amt'.

print("Total Purchase Amount for Winter Season Purchases with PayPal Payment Method: ", amt)
```

# The Average Purchase Amount for Customers who have a Subscription and used Credit Card as the Payment Method, but did not use a Promo Code

```python
pay = df[(df['Subscription Status'] == 'Yes') & (df['Payment Method'] == 'Credit Card') & (df['Promo Code Used'] == 'No')]['Purchase Amount (USD)'].mean()
print("Average Purchase Amount for Customers with Subscription, Credit Card Payment, and No Promo Code:", pay)


# The above code calculates the mean (average) purchase amount (in USD) for a specific group of customers.
# It filters the DataFrame 'df' to include only rows where certain conditions are met:
# 1. 'Subscription Status' is 'Yes' - customers with an active subscription.
# 2. 'Payment Method' is 'Credit Card' - customers who used a credit card for payment.
# 3. 'Promo Code Used' is 'No' - customers who did not use a promotional code.
# It calculates the mean (average) purchase amount for the filtered DataFrame.

print("Average Purchase Amount for Customers with Subscription, Credit Card Payment, and No Promo Code:", pay)
```

# The Most Bought Category in the Summer Season

```python
ct = df[df['Season'] == 'Summer']['Category'].mode()[0]

# Select rows from DataFrame 'df' where the 'Season' column has the value 'Summer'.
# Access the 'Category' column of the filtered DataFrame and calculate its mode.
# The mode is the most frequently occurring value in the 'Category' column.
# Extract the first mode value. In case of multiple modes, this will get the first one.
# The mode() function returns a Series, so we use [0] to get the actual mode value.

print("Most Bought Category in the Summer Season:", ct)
```
